# Capstone Project: Sentiment Analysis on Financial News

# Background

In the world of finance, timely and accurate information can be the difference between success and failure. Financial news and headlines can have a significant impact on market movements, and traders, investors, and analysts rely heavily on such news to make informed decisions. However, manually analyzing and classifying the sentiment of financial news can be a time-consuming and challenging task, especially given the vast amount of data that is available.

To overcome this challenge, there has been a growing interest in using automated sentiment analysis systems that can quickly and accurately analyze and classify the sentiment of financial news to predict market movements. Sentiment analysis involves extracting subjective information from text data, such as positive or negative opinions, emotions, and attitudes. By analyzing the sentiment of financial news, traders and investors can gain insights into market trends and make better investment decisions.

In recent years, machine learning models such as BERT (Bidirectional Encoder Representations from Transformers) and FinBERT have emerged as powerful tools for natural language processing tasks, including sentiment analysis. BERT is a pre-trained language model that uses a transformer-based architecture and has achieved state-of-the-art performance on various natural language processing tasks. FinBERT is a specialized version of BERT that is specifically trained on financial news and documents.

Given the potential of these models to improve the accuracy and efficiency of sentiment analysis for financial headlines, this project aims to develop a sentiment analysis system for financial news using BERT and FinBERT. By training and fine-tuning these models on a large dataset of financial headlines, we hope to develop a system that can accurately predict the sentiment of financial news, particularly for stock markets. Such a system could assist traders, investors, and analysts in making informed decisions and provide valuable insights into market trends.

# Problem Statement

The financial industry generates a vast amount of textual data, such as news articles, press releases, and earnings reports. Extracting sentiment from this data is crucial for making informed investment decisions. However, traditional sentiment analysis models often struggle to accurately capture the nuances and complexities of financial language. To overcome this challenge, I propose using word embedding models, BERT, and FinBERT, to improve the accuracy of sentiment analysis on financial news.

The goal is to compare the performance of these three models in capturing the sentiment of financial news articles. I will fine-tune the pretrained BERT and FinBERT models on the Kaggle dataset and evaluate their performance on the sentiment classification task. I will also compare the performance of these models to a traditional sentiment analysis model that uses word embeddings generated by Keras. By comparing the performance of these models, I aim to provide insights into the best approach for sentiment analysis on financial news data.

In short, the objective of this project is to perform sentiment analysis on financial news to classify the sentiment of the news as positive, negative or neutral. The metric used are accuracy and f1 score.

# Train Test Split

```python
def split_x_y(dataset):
    x = []
    y = []
    for headline in dataset['headlines']:
        x.append(headline)                            # Alternatively, df['headlines'].tolist()
        
    for label in dataset['sentiment']:
        if label == 'positive':
            y.append(2)
        elif label == 'neutral':
            y.append(1)
        elif label == 'negative':
            y.append(0)
    return x, y 

X, y = split_x_y(df)
```

### Function

```python
def word_embedding(checkpoint_name, classweight=False, maxlen=60, embedding_dim=100, learning_rate=0.0001, dropout_rate=0.2):
    
  # Set random seed for reproducibility
    tf.random.set_seed(22)

    # Define the vocabulary
    tokenizer = Tokenizer(oov_token='<OOV>')
    tokenizer.fit_on_texts(X_train)
    word_index = tokenizer.word_index

    #convert text to sequences
    training_sequences = tokenizer.texts_to_sequences(X_train)
    validation_sequences = tokenizer.texts_to_sequences(X_val)

    #pad the sequence to maxlength
    training_padded = pad_sequences(training_sequences,
                                    maxlen=maxlen,
                                    padding='post')
    validation_padded = pad_sequences(validation_sequences,
                                    maxlen=maxlen,
                                    padding='post')
    
    #define the model
    vocab_size = len(word_index)+1   #add 1 for the padding token '0'
    model = tf.keras.Sequential([
        tf.keras.layers.Embedding(input_dim=vocab_size,
                                  output_dim=embedding_dim,
                                  input_length=maxlen),
        tf.keras.layers.GlobalAveragePooling1D(),
        tf.keras.layers.Dense(16, activation='relu'),
        tf.keras.layers.Dropout(dropout_rate),
        tf.keras.layers.Dense(3, activation='softmax')
    ])
    
    #compile the model with optimizer, loss function and metrics
    model.compile(optimizer = tf.keras.optimizers.Adam(learning_rate=learning_rate,
                                                   beta_1=0.9,
                                                   beta_2=0.999,              
                                                   epsilon=1e-7,
                                                   amsgrad=False),
              #'sparse_categorical_crossentropy' is used if the label is not in one hot matrix, but are integers
              # no softmax activation in 'categorical_crossentropy' because it is assumed the outputs are prob
              loss = 'sparse_categorical_crossentropy',
              metrics = ['accuracy'])

    #fit the model
    early_stop = EarlyStopping(monitor='val_loss', patience=3)  #setting early stop for 3 epochs if no improvement in val loss
    checkpoint = ModelCheckpoint(checkpoint_name, monitor='val_loss', mode='min', save_best_only=True)

    if classweight == False:
        model.fit(x=training_padded, y=np.array(y_train),
                  validation_data=(validation_padded, np.array(y_val)),
                  epochs=30, batch_size=32, callbacks=[early_stop, checkpoint])

    elif classweight == True:
        model.fit(x=training_padded, y=np.array(y_train),
                  validation_data=(validation_padded, np.array(y_val)),
                  class_weight=class_weight_dict,
                  epochs=30, batch_size=32,  callbacks=[early_stop, checkpoint])
```

1) A larger vocabulary size and dimension can result in a higher computational cost and longer training time. A smaller vocabulary size and dimension can result in a loss of important information. In this case, vocab size is taken 100% from the train dataset.


2) Max pooling takes the maximum value from each feature map along the temporal dimension, while global average pooling computes the average value of each feature map along the same dimension. Max pooling is generally better suited for tasks where the exact location of the feature in the sequence is important, such as in tasks where there are strong local dependencies in the input data. On the other hand, global average pooling is better suited for tasks where the overall distribution of features is more important than their exact locations in the sequence. In text classification tasks, global average pooling is often preferred over max pooling because it can better capture the overall distribution of features in the sequence, which is often more important than the exact location of specific features. This is especially true for longer texts, where the exact position of relevant features may vary more. Hence, global average pooling is used.

# 2) BERT (bert-base-uncased)

### Showing the distribution of token frequency in each sentence using BertTokenizer

```python
def max_length(dataset):
    tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
    max_length = 0
    length = []
    for token in dataset:
        encoded = tokenizer.encode(token, add_special_tokens=False)
        if len(encoded) > max_length:
            max_length = len(encoded)
        length.append(len(encoded))
    
    plt.hist(length, bins=30)
    plt.title('Histogram for Length of Tokens')
    return max_length

max_length(X_train)
```

### Function

Learning rate 0.001 produces poor performance result. Hence, it is adjusted to 0.0001. Also, early stopping is included to avoid overfitting and poor generalization performance.

```python
def bert(checkpoint_name, maxlen=150, batch_size=32, dropout=0.2, dense=64, learning_rate=0.0001, classweight=False):

    #tokernization
    tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

    #encoding
    train_encoded = tokenizer.batch_encode_plus(X_train,
                                                padding='max_length',
                                                max_length=maxlen,              #default = 148 + 2 special tokens([CLS],[SEP])
                                                add_special_tokens=True,
                                                return_attention_mask=True,     #0 for padded token and 1 for true token
                                                return_tensors='tf')

    val_encoded = tokenizer.batch_encode_plus(X_val,
                                              padding='max_length',
                                              max_length=maxlen,               
                                              add_special_tokens=True,
                                              return_attention_mask=True,  
                                              return_tensors='tf')

    #input id
    input_id_train = train_encoded['input_ids']
    input_id_val = val_encoded['input_ids']

    #attention mask
    attention_train = train_encoded['attention_mask']
    attention_val = val_encoded['attention_mask']

    #token type id
    token_type_train = train_encoded['token_type_ids']
    token_type_val = val_encoded['token_type_ids']

    #labels
    labels_train = y_train
    labels_val = y_val

    #create standard format for tensorflow dataset objects
    def map_example_to_dict(input_ids, attention_masks, token_type_ids, label):
        return {
          "input_ids": input_ids,
          "attention_masks": attention_masks,
          "token_type_ids": token_type_ids,  
      }, label


    #create tensorflow dataset objects
    dataset_train = tf.data.Dataset.from_tensor_slices((map_example_to_dict(input_id_train,
                                                                            attention_train,
                                                                            token_type_train,
                                                                            labels_train))).shuffle(len(input_id_train)).batch(64)

    dataset_val = tf.data.Dataset.from_tensor_slices((map_example_to_dict(input_id_val,
                                                                          attention_val,
                                                                          token_type_val,
                                                                          labels_val))).batch(64)
    
    # Set random seed for reproducibility
    tf.random.set_seed(22)

    # Define the fine-tuned model
    model = TFBertForSequenceClassification.from_pretrained('bert-base-uncased', num_labels=3) 
    inputs = tf.keras.Input(shape=(maxlen,), name='input_ids', dtype=tf.int32)
    input_mask = tf.keras.Input(shape=(maxlen,), name='attention_masks', dtype=tf.int32)
    segment_ids = tf.keras.Input(shape=(maxlen,), name='token_type_ids', dtype=tf.int32)

    # Define the layers of the fine-tuned model
    bert_output = model(inputs, attention_mask=input_mask, token_type_ids=segment_ids)
    logits = bert_output.logits
    dense_layer = tf.keras.layers.Dense(dense, activation='relu')(logits)
    dropout_layer = tf.keras.layers.Dropout(dropout)(dense_layer)
    output_layer = tf.keras.layers.Dense(3, activation='softmax')(dropout_layer)

    # Define the complete model
    model = tf.keras.Model(inputs=[inputs, input_mask, segment_ids], outputs=output_layer)
    
    #compile the model with optimizer, loss function and metrics
    model.compile(optimizer = tf.keras.optimizers.Adam(learning_rate=learning_rate,
                                                   beta_1=0.9,
                                                   beta_2=0.999,              
                                                   epsilon=1e-7,
                                                   amsgrad=False),
              #'sparse_categorical_crossentropy' is used if the label is not in one hot matrix, but are integers
              # no softmax activation in 'categorical_crossentropy' because it is assumed the outputs are prob
              loss = 'sparse_categorical_crossentropy',
              metrics = ['accuracy'])
    
    #model fitting
    early_stop = EarlyStopping(monitor='val_loss', patience=3)   #setting early stop for 3 epochs if no improvement in val loss
    checkpoint = ModelCheckpoint(checkpoint_name, monitor='val_loss', mode='min', save_best_only=True)
    if classweight == False:
        model.fit(dataset_train, validation_data=dataset_val,
                  epochs=10, batch_size=batch_size, callbacks=[early_stop, checkpoint])
    elif classweight == True:
        model.fit(dataset_train, validation_data=dataset_val,
                  epochs=10, batch_size=batch_size, callbacks=[early_stop, checkpoint],
                  class_weight=class_weight_dict)
```

# Evaluation of Bert Model

```python
def cmatrix_and_creport_generation(checkpoint_name, maxlen=150):
    
    #tokernization
    tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

    #encoding
    test_encoded = tokenizer.batch_encode_plus(X_test,
                                               padding='max_length',
                                               max_length=maxlen,            
                                               add_special_tokens=True,
                                               return_attention_mask=True,  
                                               return_tensors='tf')
    
    #input id
    input_id_test = test_encoded['input_ids']
    #attention mask
    attention_test = test_encoded['attention_mask']
    #token type id
    token_type_test = test_encoded['token_type_ids']
    #labels
    labels_test = y_test

    #create standard format for tensorflow dataset objects
    def map_example_to_dict(input_ids, attention_masks, token_type_ids, label):
        return {
          "input_ids": input_ids,
          "attention_masks": attention_masks,
          "token_type_ids": token_type_ids,  
      }, label


    #create tensorflow dataset objects
    dataset_test = tf.data.Dataset.from_tensor_slices((map_example_to_dict(input_id_test,
                                                                           attention_test,
                                                                           token_type_test,
                                                                           labels_test))).batch(64)
    
    # Load the BERT model with the custom object scope
    model = tf.keras.models.load_model(checkpoint_name,
                                   custom_objects={"TFBertForSequenceClassification": transformers.TFBertForSequenceClassification})
    
    # Make predictions on test data
    y_pred = model.predict(dataset_test)
    y_true = np.concatenate([y for x, y in dataset_test], axis=0)

    # Generate confusion matrix
    cm = confusion_matrix(y_true, np.argmax(y_pred, axis=1))
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=[0,1,2])
    disp.plot()
    plt.show()

    # Generate classification report
    report = classification_report(y_true, np.argmax(y_pred, axis=1))
    print(f"Classification report:\n\n{report}")
```

# 3) FinBERT (yiyanghkust/finbert-tone)

### Showing the distribution of token frequency in each sentence using BertTokenizer

```python
#to determine max encoded sequence length to adjust max_length parameter in tokenizer
def max_length(dataset):
    tokenizer = BertTokenizer.from_pretrained('yiyanghkust/finbert-tone')
    max_length = 0
    length = []
    for token in dataset:
        encoded = tokenizer.encode(token, add_special_tokens=False)   #special_tokens=False because it is accounted for in next step
        if len(encoded) > max_length:
            max_length = len(encoded)
        length.append(len(encoded))
        
    plt.hist(length, bins=30)
    plt.title('Histogram for Length of Tokens')
    return max_length

max_length(X_train)
```

### Function
1) for model finbert 1,2 and 3, maxlen=134, batch_size=64, dense=64, dropout=0.2
2) for model finbert 1.1, 2.1 and 3.1, maxlen=80, batch_size=32, dense=32, dropout=0.3

```python
def finbert(checkpoint_name, maxlen=134, batch_size=64, dropout=0.2, dense=64, learning_rate=0.0001, classweight=False):

    #tokernization
    tokenizer = BertTokenizer.from_pretrained('yiyanghkust/finbert-tone')

    #encoding
    train_encoded = tokenizer.batch_encode_plus(X_train,
                                                padding='max_length',
                                                max_length=maxlen,              #default 134=132 + 2 special tokens([CLS],[SEP])
                                                add_special_tokens=True,
                                                truncation=True,
                                                return_attention_mask=True,     #0 for padded token and 1 for true token
                                                return_tensors='tf')

    val_encoded = tokenizer.batch_encode_plus(X_val,
                                              padding='max_length',
                                              max_length=maxlen,               
                                              add_special_tokens=True,
                                              truncation=True,
                                              return_attention_mask=True,  
                                              return_tensors='tf')

    #input id
    input_id_train = train_encoded['input_ids']
    input_id_val = val_encoded['input_ids']

    #attention mask
    attention_train = train_encoded['attention_mask']
    attention_val = val_encoded['attention_mask']

    #token type id
    token_type_train = train_encoded['token_type_ids']
    token_type_val = val_encoded['token_type_ids']

    #labels
    labels_train = y_train
    labels_val = y_val

    #create standard format for input
    def map_example_to_dict(input_ids, attention_masks, token_type_ids, label):
        return {
          "input_ids": input_ids,
          "attention_masks": attention_masks,
          "token_type_ids": token_type_ids, 
      }, label

    #create tensorflow dataset objects
    dataset_train = tf.data.Dataset.from_tensor_slices((map_example_to_dict(input_id_train,
                                                                            attention_train,
                                                                            token_type_train,
                                                                            labels_train))).shuffle(len(input_id_train)).batch(64)

    dataset_val = tf.data.Dataset.from_tensor_slices((map_example_to_dict(input_id_val,
                                                                          attention_val,
                                                                          token_type_val,
                                                                          labels_val))).batch(64)

    #fine tuning
    # Set random seed for reproducibility
    tf.random.set_seed(22)

    # Define the fine-tuned model
    #ipuneetrathore/bert-base-cased-finetuned-finBERT, yiyanghkust/finbert-tone, ProsusAI are trained on different datasets
    #yiyanghkust/finbert-tone is more popular
    model = TFBertForSequenceClassification.from_pretrained('yiyanghkust/finbert-tone',num_labels=3) 
    inputs = tf.keras.Input(shape=(maxlen,), name='input_ids', dtype=tf.int32)
    input_mask = tf.keras.Input(shape=(maxlen,), name='attention_masks', dtype=tf.int32)
    segment_ids = tf.keras.Input(shape=(maxlen,), name='token_type_ids', dtype=tf.int32)

    # Define the layers of the fine-tuned model
    bert_output = model(inputs, attention_mask=input_mask, token_type_ids=segment_ids)
    logits = bert_output.logits
    dense_layer = tf.keras.layers.Dense(dense, activation='relu')(logits)
    dropout_layer = tf.keras.layers.Dropout(dropout)(dense_layer)
    output_layer = tf.keras.layers.Dense(3, activation='softmax')(dropout_layer)

    # Define the complete model
    model = tf.keras.Model(inputs=[inputs, input_mask, segment_ids], outputs=output_layer)
    
    #compile the model with optimizer, loss function and metrics
    model.compile(optimizer = tf.keras.optimizers.Adam(learning_rate=learning_rate,       
                                                   beta_1=0.9,                 
                                                   beta_2=0.999,               #optimizer will consider the last 1/(1-0.999) = 1000 timesteps when computing the second moment estimates.
                                                   epsilon=1e-7,               #A small floating point value to prevent division by zero in the optimizer's update step
                                                   amsgrad=False),
              #'sparse_categorical_crossentropy' is used if the label is not in one hot matrix, but are integers
              # no softmax activation in 'categorical_crossentropy' because it is assumed the outputs are prob
              loss = 'sparse_categorical_crossentropy',
              metrics = ['accuracy'])
    
     
    #model fitting
    early_stop = EarlyStopping(monitor='val_loss', patience=3)   #setting early stop for 3 epochs if no improvement in val loss
    checkpoint = ModelCheckpoint(checkpoint_name, monitor='val_loss', mode='min', save_best_only=True)
    if classweight == False:
        model.fit(dataset_train, validation_data=dataset_val,
                  epochs=10, batch_size=batch_size, callbacks=[early_stop, checkpoint])
    elif classweight == True:
        model.fit(dataset_train, validation_data=dataset_val,
                  epochs=10, batch_size=batch_size, callbacks=[early_stop, checkpoint],
                  class_weight=class_weight_dict)
```

Note:

Beta_1 is the exponential decay rate for the first moment estimates, which corresponds to the moving average of the gradient. A value close to 1.0 will give the optimizer more weight to the past gradients, which may be helpful in some cases, but it can also result in slower convergence. A value closer to 0.0 will give less weight to the past gradients, which may result in faster convergence, but it can also lead to oscillations and instability.

Beta_2 is the exponential decay rate for the second moment estimates, which corresponds to the moving average of the gradient squared. A value close to 1.0 will give the optimizer more weight to the past squared gradients, which may be helpful in some cases, but it can also result in slower convergence. A value closer to 0.0 will give less weight to the past squared gradients, which may result in faster convergence, but it can also lead to oscillations and instability.

# Evaluation of FinBert Model

```python
def cmatrix_and_creport_generation(checkpoint_name, maxlen=134):
    
    #tokernization
    tokenizer = BertTokenizer.from_pretrained('yiyanghkust/finbert-tone')
    test_encoded = tokenizer.batch_encode_plus(X_test,
                                               padding='max_length',
                                               max_length=maxlen,            
                                               add_special_tokens=True,
                                               truncation=True,
                                               return_attention_mask=True,  
                                               return_tensors='tf')

    #input id
    input_id_test = test_encoded['input_ids']
    #attention mask
    attention_test = test_encoded['attention_mask']
    #token type id
    token_type_test = test_encoded['token_type_ids']
    #labels
    labels_test = y_test

    #create standard format for input
    def map_example_to_dict(input_ids, attention_masks, token_type_ids, label):
        return {
          "input_ids": input_ids,
          "attention_masks": attention_masks,
          "token_type_ids": token_type_ids, 
      }, label

    #create tensorflow dataset objects
    dataset_test = tf.data.Dataset.from_tensor_slices((map_example_to_dict(input_id_test,
                                                                           attention_test,
                                                                           token_type_test,
                                                                           labels_test))).batch(64)
    
    # Load the BERT model with the custom object scope
    model = tf.keras.models.load_model(checkpoint_name,
                                   custom_objects={"TFBertForSequenceClassification": transformers.TFBertForSequenceClassification})
    
    # Make predictions on test data
    y_pred = model.predict(dataset_test)
    y_true = np.concatenate([y for x, y in dataset_test], axis=0)

    # Generate confusion matrix
    cm = confusion_matrix(y_true, np.argmax(y_pred, axis=1))
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=[0,1,2])
    disp.plot()
    plt.show()

    # Generate classification report
    report = classification_report(y_true, np.argmax(y_pred, axis=1))
    print(f"Classification report:\n\n{report}")
```

```python
models = ['finbert1', 'finbert2', 'finbert3']

for m in models:
    print(f'Model {m}:\n')
    cmatrix_and_creport_generation(checkpoint_name=f'model_checkpoint/{m}.h5', maxlen=134)
    print(f'\n\n')
```

# Conclusion

Word embedding is a technique that represents words as numerical vectors, which can be used to capture the semantic relationships between words. However, it does not capture the context in which words are used, which leads to inaccurate sentiment analysis.

BERT, on the other hand, is a pre-trained language model that uses a bidirectional transformer architecture to capture the context in which words are used. However, BERT is computationally expensive to fine-tune the model.

FinBERT is a pre-trained language model that has been specifically designed for financial sentiment analysis. It is based on the BERT architecture and has been fine-tuned on financial news articles and social media posts. 

In this project, FinBERT has been shown to achieve higher accuracy in financial sentiment analysis than the general-purpose language model BERT, with reported accuracy rates of up to 85% and weighted average F1 score of up to 84%.
